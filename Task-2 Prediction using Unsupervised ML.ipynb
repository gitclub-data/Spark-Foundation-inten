```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv("dataset/Iris.csv")
```

```python
df.head()
```

```python
df.drop('Id',inplace=True ,axis=1)
```

```python
df.head()
```

```python
df['Species'].unique()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
df.groupby('Species').max().plot.bar(ax=ax[0],title='max in Cm')
df.groupby('Species').min().plot.bar(ax=ax[1],title='min in Cm')
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
sns.scatterplot('SepalLengthCm','SepalWidthCm',data=df,hue='Species',ax=ax[0])
sns.scatterplot('SepalLengthCm','PetalLengthCm',data=df,hue='Species',ax=ax[1])
sns.scatterplot('SepalLengthCm','PetalWidthCm',data=df,hue='Species',ax=ax[2])
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
sns.scatterplot('SepalWidthCm','PetalLengthCm',data=df,hue='Species',ax=ax[0])
sns.scatterplot('SepalWidthCm','PetalWidthCm',data=df,hue='Species',ax=ax[1])
sns.scatterplot('PetalLengthCm','PetalWidthCm',data=df,hue='Species',ax=ax[2])
```

```python
#K Means Anaysis and clustring
data=df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].values
```

```python
Model = KMeans(n_clusters=3)
Model.fit(data)   
```

```python
centroids=Model.cluster_centers_
centroids
```

```python
pred=Model.fit_predict(data)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
sns.scatterplot('SepalLengthCm','SepalWidthCm',data=df,hue='Species',ax=ax[0])
plt.scatter(data[:,0],data[:,1],c=pred,cmap='rainbow')
plt.scatter(centroids[:,0],centroids[:,1],color="black")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
sns.scatterplot('SepalLengthCm','PetalLengthCm',data=df,hue='Species',ax=ax[0])
plt.scatter(data[:,0],data[:,2],c=pred,cmap='rainbow')
plt.scatter(centroids[:,0],centroids[:,2],color="black")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
sns.scatterplot('SepalLengthCm','PetalWidthCm',data=df,hue='Species',ax=ax[0])
plt.scatter(data[:,0],data[:,3],c=pred,cmap='rainbow')
plt.scatter(centroids[:,0],centroids[:,3],color="black")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
sns.scatterplot('SepalWidthCm','PetalLengthCm',data=df,hue='Species',ax=ax[0])
plt.scatter(data[:,1],data[:,2],c=pred,cmap='rainbow')
plt.scatter(centroids[:,1],centroids[:,2],color="black")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
sns.scatterplot('SepalWidthCm','PetalWidthCm',data=df,hue='Species',ax=ax[0])
plt.scatter(data[:,1],data[:,3],c=pred,cmap='rainbow')
plt.scatter(centroids[:,1],centroids[:,3],color="black")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
sns.scatterplot('PetalLengthCm','PetalWidthCm',data=df,hue='Species',ax=ax[0])
plt.scatter(data[:,2],data[:,3],c=pred,cmap='rainbow')
plt.scatter(centroids[:,2],centroids[:,3],color="black")
```
